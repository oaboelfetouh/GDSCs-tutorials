```python
# add HP, Attack, Defense ... together! #WAY1
df['total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] +df['Speed']
df.head()
```

```python
# add HP, Attack, Defense ... together! #WAY2
df['total'] = df.iloc[:, 3:9].sum(axis = 1)
df.head()
```

```python
df.columns # returning sth called pandas.core.indexes.base.Index
df.columns.values # returning an array 
list(df.columns.values) #returns a list! - and this is useful - 
```

```python
cols = list(df.columns.values)

#this method is aslo used for re-ordering the dataframe
df = df[cols[0:4] + [cols[-1]] ]  #lma t + single column lazem t5leh list!
df.head()
```
