```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# Plot and Scatter

```python
x = list(range(-10,11))
y = [i**2 for i in x]
```

```python
#very basic plot
plt.plot(x,y)
plt.show()
```

```python
#very basic scatter plot
plt.scatter(x,y)
plt.show()
```

```python
#add more complexity 
plt.plot(x,y,
         marker = 'o' ,# markers can be : o , * , x , + , and many other things
         linestyle = 'dotted' ,# linestyle can be : 'solid' , 'dotted' , ' dashed', and other
         color = 'r' ,# colot can be : 'r' , 'b', or any color colde "use google"
         linewidth = '1.5', 
         )

plt.xlim(0,10) # to set the x range
plt.ylim(0,100) # to set the y range


#Some Decorations
plt.title('Hi There, Iam using Whatsapp')
plt.xlabel("hi x")
plt.ylabel("hi y")

plt.grid() 
# you can use grid(axis = 'y') or grid(axis = 'x'), also you can add some decorations
plt.show()
```

```python
#NOTE
plt.plot([1,2,3,4,5,6]) # the x axis is the indix :))
```

```python
#NOTE 
plt.plot(x,y)
plt.plot(list(range(0,100)))
plt.show()

# we can make graphs in the same plot!
```

# Histogram and bar chart

```python
bar_data_X = ['a', 'b', 'c']
bar_data_Y = [50,60,30]
hist_data = [50,50,50,50,50, 20,20,20,10, 30,30,30,30,30,30]
```

```python
#simple bar chart
plt.bar(bar_data_X, bar_data_Y
        , color = 'r'
        , width = 0.3)

plt.show()
```

```python
plt.barh(bar_data_X, bar_data_Y, color ='b', height = 0.3)
plt.show()
```

```python
#simple histogram
plt.hist(hist_data)
plt.show()
```
